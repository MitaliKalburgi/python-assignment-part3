```python
# Part B - Read & Search

try:
    # Step 1: Read file
    with open("python_notes.txt", "r", encoding="utf-8") as file:
        lines = file.readlines()

    # Step 2: Count & Print total number of lines
    print("\nFile Contents:\n")

    for index, line in enumerate(lines, start=1):
        print(f"{index}. {line.strip()}")

    total_lines = len(lines)
    print(f"\nTotal number of lines: {total_lines}")

    # Step 3: Keyword search
    keyword = input("\nEnter a keyword to search: ").lower()

    found = False

    for line in lines:
        if keyword in line.lower():
            print(line.strip())                # print all lines that contain that keyword
            found = True

    if not found:
        print("No matching lines found.")

except FileNotFoundError:
    print("Error: File not found. Run Part A first.")

except IOError as e:
    print(f"File error: {e}")
```

```python
# Step 1 - Fetch Products

url = "https://dummyjson.com/products?limit=20"

try:
    response = requests.get(url)                     # GET request
    response.raise_for_status()                      # Check for errors
    data = response.json()

    products = data["products"]                      # parse the JSON response

    # Print table header
    print(f"{'ID':<4}| {'Title':<45} | {'Category':<13} | {'Price':<8} | {'Rating'}")
    print(f"{'':->4}|{'':->47}|{'':->15}|{'':->10}|{'':->7}")

    # Print product data
    for p in products:
        print(f"{p['id']:<4}| {p['title']:<45} | {p['category']:<13} | ${p['price']:<7.2f} | {p['rating']}")

except requests.exceptions.ConnectionError:
    print("Error: No internet connection.")

except requests.exceptions.HTTPError as e:
    print(f"HTTP Error: {e}")

except requests.exceptions.RequestException as e:
    print(f"Error: {e}")
```

```python
# Step 2 - Filter and Sort

# Filter products with rating >= 4.5
filtered_products = []

for p in products:
    if p["rating"] >= 4.5:
        filtered_products.append(p)

# Sort by price (descending)
filtered_products.sort(key=lambda x: x["price"], reverse=True)

# Print the filtered and sorted list.
print("\nFiltered & Sorted Products:\n")

print(f"{'ID':<4}| {'Title':<45} | {'Price':<9} | {'Rating':<6}")
print(f"{'':->4}|{'':->47}|{'':->11}|{'':->8}")

for p in filtered_products:
    print(f"{p['id']:<4}| {p['title']:<45} | ${p['price']:<8.2f} | {p['rating']:<6}")
```

**Task 3 - Exception Handling**

```python
# Part A - Guarded Calculator

def safe_divide(a, b):
    try:
        return a / b          # returns the result of a/b

    except ZeroDivisionError:
        return "Error: Cannot divide by zero"

    except TypeError:
        return "Error: Invalid input types"

# Test cases
print(safe_divide(10, 2))      # Normal case
print(safe_divide(10, 0))      # Division by zero
print(safe_divide("ten", 2))   # Invalid type
```

```python
# Part B — Guarded File Reader

def read_file_safe(filename):
    try:
        # Try to open and read file
        with open(filename, "r", encoding="utf-8") as file:
            content = file.read()
            return content

    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")

    finally:
        print("File operation attempt complete.")


# Test cases
print(read_file_safe("python_notes.txt"))   # Should work
print(read_file_safe("ghost_file.txt"))     # Should fail
```

```python
# Part C - Robust API Calls

import requests

# STEP 1 - Fetch Products
print("\n--- Fetch Products ---\n")

url = "https://dummyjson.com/products?limit=20"

try:
    response = requests.get(url, timeout=5)
    response.raise_for_status()

    data = response.json()
    products = data["products"]

    for p in products:
        print(p["title"])

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Error:", e)

# STEP 3 - Laptops Category
print("\n--- Laptops ---\n")

url = "https://dummyjson.com/products/category/laptops"

try:
    response = requests.get(url, timeout=5)
    response.raise_for_status()

    data = response.json()
    laptops = data["products"]

    for p in laptops:
        print(f"{p['title']} - ${p['price']}")

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Error:", e)

# STEP 4 — POST Request
print("\n--- POST Request ---\n")

url = "https://dummyjson.com/products/add"

new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

try:
    response = requests.post(url, json=new_product, timeout=5)
    response.raise_for_status()

    print(response.json())

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Error:", e)
```

```python
# Part D - Input Validation Loop

import requests

while True:
    user_input = input("\nEnter a product ID to look up (1–100), or 'quit' to exit: ")

    # Exit condition
    if user_input.lower() == "quit":
        print("Exiting program.")
        break

    # Validate integer input
    if not user_input.isdigit():
        print("Invalid input! Please enter a number between 1 and 100.")
        continue

    product_id = int(user_input)

    # Validate range
    if product_id < 1 or product_id > 100:
        print("Invalid range! Enter a number between 1 and 100.")
        continue

    # API call
    url = f"https://dummyjson.com/products/{product_id}"

    try:
        response = requests.get(url, timeout=5)       # GET request

        # Check status code
        if response.status_code == 404:
            print("Product not found.")

        elif response.status_code == 200:
            data = response.json()
            print(f"Product: {data['title']} - ${data['price']}")

        else:
            print("Unexpected response:", response.status_code)

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")

    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")

    except Exception as e:
        print("Error:", e)
```

**Task 4 - Logging to File**

```python
# Build a simple error logger

import requests
from datetime import datetime

# Function to write logs
def log_error(context, error_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    with open("error_log.txt", "a", encoding="utf-8") as file:
        file.write(f"[{timestamp}] ERROR in {context}: {error_type} — {message}\n")

# Intentionally trigger at least two logged entries

# 1. Trigger ConnectionError
try:
    bad_url = "https://this-host-does-not-exist-xyz.com/api"
    response = requests.get(bad_url, timeout=5)

except requests.exceptions.ConnectionError as e:
    print("Connection error triggered.")
    log_error("fetch_products", "ConnectionError", str(e))


# 2. Trigger HTTP 404 Error
try:
    url = "https://dummyjson.com/products/999"  # Non-existent product
    response = requests.get(url, timeout=5)

    if response.status_code != 200:
        print("HTTP error triggered.")
        log_error(
            "lookup_product",
            "HTTPError",
            f"{response.status_code} Not Found for product ID 999"
        )

except Exception as e:
    log_error("lookup_product", "UnexpectedError", str(e))


# Read and Print Log File
print("\n--- Error Log Contents ---\n")

try:
    with open("error_log.txt", "r", encoding="utf-8") as file:
        print(file.read())

except FileNotFoundError:
    print("No log file found.")
```
